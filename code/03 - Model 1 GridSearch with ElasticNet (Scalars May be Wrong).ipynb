## Model Tuning

```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

import patsy
import statsmodels.api as sm

from sklearn import metrics
from sklearn.metrics import r2_score, recall_score, make_scorer, f1_score, mean_squared_error

from sklearn import linear_model
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet, RidgeCV, LassoCV, ElasticNetCV

from sklearn.model_selection import train_test_split, cross_val_score, cross_val_predict, GridSearchCV , RandomizedSearchCV
from sklearn.feature_selection import SelectKBest, VarianceThreshold, f_regression

from sklearn.preprocessing import StandardScaler, PolynomialFeatures #PowerTransformer

from sklearn.neighbors import KNeighborsRegressor


%config InlineBackend.figure_format = 'retina'
%matplotlib inline

pd.set_option('max_columns',300)
```

```python
X = pd.read_csv('../datasets/data_clean_final.csv')
```

```python
X.head()
```

```python
y = X['SalePrice']
```

```python
X.drop('SalePrice', axis=1, inplace=True)
```

### Basic Linear Regression, No Holdout

```python
lr = LinearRegression()
```

```python
cross_val_score(lr, X, y, cv=5).mean()
```

### Model Tuning:  Basic Linear Regression

```python
X_train, X_test, y_train, y_test = train_test_split(X,y)

lr_train = LinearRegression()

cross_val_score(lr_train, X_train, y_train, cv=5).mean()
```

```python
lr_train.fit(X_train, y_train)

lr_train.score(X_test, y_test)
```

### Model Tuning:  Elastic Net

```python
X_train, X_test, y_train, y_test = train_test_split(X,y)

ss = StandardScaler()
X_train_sc = ss.fit_transform(X_train)
X_test_sc = ss.transform(X_test)

#Ran an initial logspace of (0.5,2,50), then narrowed.
enet_alphas = np.logspace(-1,1,20)
enet_ratio = 0.5
enet_model = ElasticNetCV(alphas=enet_alphas, l1_ratio=enet_ratio, cv=5, tol=0.0001, max_iter=1000)

enet_model = enet_model.fit(X_train_sc, y_train)

enet_model_preds_train = enet_model.predict(X_train_sc)
enet_model_preds_test = enet_model.predict(X_test_sc)

print(r2_score(y_train, enet_model_preds_train))
print(r2_score(y_test, enet_model_preds_test))  
print(enet_model.alpha_)
```

```python
preds = enet_model.predict(X_test_sc)
plt.figure(figsize=(10,8))
plt.scatter(preds,y_test,marker = '+')
plt.xlabel('Predicted values')
plt.ylabel('Actual values')
plt.plot([0,np.max(y_test)],[0,np.max(y_test)], c = 'k');
```

```python
plt.hist(y_test - preds, bins=30)
```

```python
pd.DataFrame(enet_model.coef_, index=X_train.columns).sort_values(0)
```

### Model Tuning: GridSearch with ElasticNet

```python
X_train, X_test, y_train, y_test = train_test_split(X,y)

ss = StandardScaler()
X_train_sc = ss.fit_transform(X_train)
X_test_sc = ss.transform(X_test)

param_grid_1 = {
    'alpha':[.3,.5,.6,.7,1,5],
    'l1_ratio':[0,.3,.5,.7,1]
}

gs_simple = GridSearchCV(ElasticNet(),param_grid_1,cv=5,verbose=1)

gs_simple.fit(X_train_sc,y_train)  
```

```python
gs_simple.best_estimator_
```

```python
gs_simple.best_score_
```

```python
pd.DataFrame(gs_simple.best_estimator_.coef_,index=X_train.columns).sort_values(0)
```

```python
gs_simple.best_estimator_.score(X_test_sc,y_test)
```

```python
#Did some testing with the model here and it actually performs better than this.  Varies between 0.85 and 0.92 mostly.
X_train, X_test, y_train, y_test = train_test_split(X,y)

ss = StandardScaler()
X_train_sc = ss.fit_transform(X_train)
X_test_sc = ss.transform(X_test)

gs_simple.best_estimator_.score(X_test_sc,y_test)
```
